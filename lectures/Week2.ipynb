```python
import networkx as nx

#a. 
Ga = nx.Graph()

#Two nodes have three connections and two nodes have two connections
Ga.add_edges_from([(1,2),(1,3),(1,4),(2,3),(3,4)])

#Plot the graph
nx.draw(Ga, with_labels=True)

#test if eulerian path is possible
nx.has_eulerian_path(Ga)



```

```python
import networkx as nx

G = nx.Graph()

#Undirected graph of 6 nodes and 7 links.

G.add_edges_from([(1,2),(1,3),(1,4),(2,3),(2,1),(2,4),(3,6),(1,6)])
#nx.draw(G, with_labels=True)

#The corresponding adjacency matrix.
print(nx.adjacency_matrix(G).todense())

#The corresponding link lists.
print(G.edges())

#Determine the average clustering coefficient of the network shown in
print(nx.average_clustering(G))

#Directed graph of 6 nodes and 8 directed links.
F = nx.DiGraph()

F.add_edges_from([(1,2),(1,3),(2,3),(2,4),(3,2),(3,1),(4,1),(6,1),(6,3)])

nx.draw(F, with_labels=True)

#The corresponding adjacency matrix.
print(nx.adjacency_matrix(F).todense())

#The corresponding link lists.
print(F.edges())

#Determine the average clustering coefficient of the network shown in
print(nx.average_clustering(F))

#In the (a) network, how many paths (with possible repetition of nodes and links) of length 3 exist starting from node 1 and ending at node 3? And in (b)?

#Create a function that counts the number of paths in the graph
def count_paths(G, start, end, length):
    count = 0
    if length == 0:
        if start == end:
            return 1
        else:
            return 0
    else:
        for node in G.neighbors(start):
            count += count_paths(G, node, end, length-1)
        return count

#Test the function
print(count_paths(G,1,3,3))
print(nx.all_simple_edge_paths(G,1,3,3))

#Test the function
print(count_paths(F,1,3,3))
print(nx.all_simple_edge_paths(F,1,3,3))

#With the help of a computer, count the number of cycles of length 4 in both networks.
print(nx.cycle_basis(G))
print(nx.cycle_graph(F))
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

#Generate 5000 random data with mean 2 and standard deviation 0.125

data = np.random.normal(2, 0.125, 5000)

np.histogram(data, bins=10)

#Plot the histogram
#plt.hist(data, bins=10)

#plt.plot(data)

#Generate 5000 random integers from a Poisson distribution with rate parameter 10 using numpy

numbers = np.random.poisson(10, 5000)

#print the maximum and minimum values of poisson
max = np.max(numbers)
min = np.min(numbers)

#Create a vector  𝑣 that results in a binning that puts each integer value from the poisson distribution in its own bin and where the first bin contains the minimum number you found above, and the last bin contains the maximum number

v = np.arange(min, max)

counts, bins = np.histogram(numbers, bins=v)

print("Sum over bin counts:", counts.sum())

print(f"The first element of the binning vector is {bins[0]} and the last is {bins[-1]}")
print(f"This means the first bin starts just before the minimum value of {min} and the last bin ends just after the maximum value of {max}.")

plt.bar(bins[:-1], counts, width=0.8, align='center')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Poisson distributed data')
plt.show()
```

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

# 1. Find p
N = 5000
k_avg = 10
p = k_avg / (N - 1)

# 2. Create ER network
G = nx.erdos_renyi_graph(N, p)

# 3. Extract degree distribution
degrees = [G.degree(n) for n in G.nodes()]

# 4. Binning and plotting the degree distribution
data_min = np.min(degrees)
data_max = np.max(degrees)
v = np.arange(data_min, data_max + 2) - 0.5

counts, bins = np.histogram(degrees, bins=v)

# Plot
plt.bar(bins[:-1], counts, width=0.8, align='center')
plt.xlabel('Degree')
plt.ylabel('Number of Nodes')
plt.title('Degree Distribution of ER Network')
plt.show()
```
