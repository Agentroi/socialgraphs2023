    ... some text [[Aristotle]] some more text ...

```python
import urllib.request
import re
import networkx as nx
import pandas as pd
import os
#from pyvis.network import Network


urlWest = 'https://en.wikipedia.org/wiki/Category:West_Coast_hip_hop_musicians'
response = urllib.request.urlopen(urlWest)
data = response.read()      # a `bytes` object
textWest = data.decode('utf-8')
westUrl = re.findall('href="/wiki/([^:#"]+)"', textWest)
urlWest = 'https://en.wikipedia.org/w/index.php?title=Category:West_Coast_hip_hop_musicians&pagefrom=McFedries%2C+Trevor%0ATrevor+McFedries#mw-pages'
response = urllib.request.urlopen(urlWest)
data = response.read()      # a `bytes` object
textWest = data.decode('utf-8')
westUrl2 = re.findall('href="/wiki/([^:#"]+)"', textWest)
westUrl = westUrl[17:] + westUrl2[17:]


urlEast = 'https://en.wikipedia.org/wiki/Category:East_Coast_hip_hop_musicians'
response = urllib.request.urlopen(urlEast)
data = response.read()      # a `bytes` object
textEast = data.decode('utf-8')
eastUrl = re.findall('href="/wiki/([^:#"]+)"', textEast)
urlEast = 'https://en.wikipedia.org/w/index.php?title=Category:East_Coast_hip_hop_musicians&pagefrom=Junglepussy%0AJunglepussy#mw-pages'
response = urllib.request.urlopen(urlEast)
data = response.read()      # a `bytes` object
textWest = data.decode('utf-8')
eastUrl2 = re.findall('href="/wiki/([^:#"]+)"', textWest)
urlEast = 'https://en.wikipedia.org/w/index.php?title=Category:East_Coast_hip_hop_musicians&pagefrom=Stezo#mw-pages'
response = urllib.request.urlopen(urlEast)
data = response.read()      # a `bytes` object
textWest = data.decode('utf-8')
eastUrl3 = re.findall('href="/wiki/([^:#"]+)"', textWest)
eastUrl = eastUrl[15:] + eastUrl2[15:] + eastUrl3[15:]


eastCSV = pd.read_csv('/Users/roidalsgard/Downloads/EastCoastRappers.csv')
eastNames = []
#Save all rappernames with header WikipediaPageName in column 2 and remove everything inside brackets
for rapper in eastCSV['WikipediaPageName']:
    rapper = re.sub(r'\([^)]*\)', '', rapper)
    eastNames.append(rapper)

westCSV = pd.read_csv('/Users/roidalsgard/Downloads/WestCoastRappers.csv')
westNames = []
#Save all rappernames with header WikipediaPageName in column 2 and remove everything inside brackets
for rapper in westCSV['WikipediaPageName']:
    rapper = re.sub(r'\([^)]*\)', '', rapper)
    westNames.append(rapper)

#Join the two lists with westNames in column 0 and westUrl in column 1 and add 'en.wikipedia.org/wiki/' to the fron of each url
west = pd.DataFrame({'Name': westNames, 'URL': westUrl})
east = pd.DataFrame({'Name': eastNames, 'URL': eastUrl})

#Create a graph
G = nx.DiGraph()

G.add_node(east.URL[0])

print(east.URL)

for file in east.URL:
    #color red
    text = open('/Users/roidalsgard/Documents/GitHub/socialgraphs2023/wikis/east/' + file + '.txt', 'r')
    text = text.read()
    G.add_node(file, attr_coast = 'east', attr_length = len(text), color='red')
    wikiLinks = re.findall('href="/wiki/([^:#"]+)"', text)
    for link in wikiLinks:
        #skip if the link is to itself
        if link == file:
            continue
        if link in east.URL.values:
            G.add_edge(file, link)
        if link in west.URL.values:
            G.add_edge(file, link)

for file in west.URL:
    text = open('/Users/roidalsgard/Documents/GitHub/socialgraphs2023/wikis/west/' + file + '.txt', 'r')
    text = text.read()
    G.add_node(file, attr_coast = 'west', attr_length = len(text), color='green')
    wikiLinks = re.findall('href="/wiki/([^:#"]+)"', text)
    for link in wikiLinks:
        #skip if the link is to itself
        if link == file:
            continue
        if link in east.URL.values:
            G.add_edge(file, link)
        if link in west.URL.values:
            G.add_edge(file, link)

#Print all notes including their attributes



print(len(east))
print(len(west))
print(len(nx.nodes(G)))
print(len(nx.edges(G)))

#Plot the in and out-degree distributions. What do you observe? Can you explain why the in-degree distribution is different from the out-degree distribution?
inDegree = []
outDegree = []
for node in G.nodes():
    inDegree.append(G.in_degree(node))
    outDegree.append(G.out_degree(node))

import matplotlib.pyplot as plt
plt.hist(inDegree, bins=100)
plt.show()
plt.hist(outDegree, bins=100)
plt.show()

#Find the 10 most popular rappers (the nodes with highest out-degree). Are there any West Coast rappers in this list? How many? Why do you think that is?
outDegree = []
for node in G.nodes():
    outDegree.append(G.out_degree(node))

outDegree = pd.DataFrame({'Name': G.nodes(), 'OutDegree': outDegree})
outDegree = outDegree.sort_values(by=['OutDegree'], ascending=False)
print(outDegree.head(6))

#Find the 10 most popular rappers (the nodes with highest in-degree). Are there any West Coast rappers in this list? How many? Why do you think that is?
inDegree = []
for node in G.nodes():
    inDegree.append(G.in_degree(node))

inDegree = pd.DataFrame({'Name': G.nodes(), 'InDegree': inDegree})
inDegree = inDegree.sort_values(by=['InDegree'], ascending=False)
print(inDegree.head(6))

#Who are the top 5 most connected West coast rappers (again in terms of both in/out-degree)?
westDegree = []
for node in west.URL:
    westDegree.append(G.degree(node))

westDegree = pd.DataFrame({'Name': west.URL, 'Degree': westDegree})
westDegree = westDegree.sort_values(by=['Degree'], ascending=False)
print(westDegree.head(5))

#Who are the top 5 most connected East coast rappers (again in terms of both in/out-degree)?
eastDegree = []
for node in east.URL:
    eastDegree.append(G.degree(node))

eastDegree = pd.DataFrame({'Name': east.URL, 'Degree': eastDegree})
eastDegree = eastDegree.sort_values(by=['Degree'], ascending=False)
print(eastDegree.head(5))


#plot the graph with small nodes and labels and isolated nodes removed
#pos = nx.spring_layout(G)
#nx.draw(G, pos, node_size=10, with_labels=True, font_size=4)

#Calculate the percentage of edges going to respectively west and east coast rappers for the top 5 nodes of west and east

# Calculate total degree (in-degree + out-degree) for each node
total_degrees = {node: G.in_degree(node) + G.out_degree(node) for node in G.nodes()}

# Group nodes by their total degree
degree_values = sorted(set(total_degrees.values()))
histogram = [list(total_degrees.values()).count(i) for i in degree_values]

# Plot the degree distribution in log-log scale
plt.figure(figsize=(10, 6))
plt.scatter(degree_values, histogram, color='b')
plt.xscale('log')
plt.yscale('log')
plt.title("Log-Log Degree Distribution")
plt.ylabel("Number of Nodes (log scale)")
plt.xlabel("Degree (log scale)")
plt.grid(which="both", linestyle="--")
plt.show()



filename = "/Users/roidalsgard/Documents/GitHub/socialgraphs2023/lectures/largest_component.graphml"
nx.write_graphml(G, filename)


# nt = Network(notebook=True)
# nt.from_nx(G)
# nt.set_options("""
# {
#   "physics": {
#     "barnesHut": {
#       "gravitationalConstant": -2000,
#       "centralGravity": 0.1,
#       "springLength": 200,
#       "springConstant": 0.01,
#       "damping": 0.3,
#       "avoidOverlap": 0.1
#     },
#     "minVelocity": 0.1,
#     "maxVelocity": 50
#   }
# }
# """)
# nt.show('nx.html')
```
